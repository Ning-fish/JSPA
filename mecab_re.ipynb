## 定義字典斷詞方法  
  mecab_parse 回傳dict_type字典形式日文句的詞清單，每詞包含word,kana,lemma,pos,itype,iform六欄位  
  mecab_parse_space 回傳空格斷詞結果，詞之間以空白隔開  
  mecab_parse_pos 回傳詞性斷詞結果，詞之間以行隔開，每詞包含word,kana,lemma,pos,itype,iform六詞性欄位


```
def mecab_parse(sentence, dict_type='unidic'):
  wdict_list =[]

  if dict_type=='ipadic':
    seg_sentence = mecab.parse(sentence)
    word_list = seg_sentence.split('\n')
  
    for seg_word in word_list:
      fields = seg_word.split('\t')
      if(len(fields) < 4):
        continue

      word = fields[0]
      kana = fields[1]
      lemma = fields[2]
      pos = fields[3]
      itype = fields[4]   # inflection type
      iform = fields[5]   # inflection form

      wdict = {'word':word, 'kana':kana, 'lemma':lemma, 'pos':pos, 'itype':itype, 'iform':iform}
      wdict_list.append(wdict)
  elif dict_type=='unidic':
    word_list = unidic_parse(sentence)
    for seg_word in word_list:
      word = str(seg_word)
      kana = seg_word.feature.kana
      lemma = seg_word.feature.lemma
      pos  = seg_word.pos
      itype = seg_word.feature.cType
      iform = seg_word.feature.cForm
 
      wdict = {'word':word, 'kana':kana, 'lemma':lemma, 'pos':pos, 'itype':itype, 'iform':iform}
      wdict_list.append(wdict)

  return wdict_list

def mecab_parse_space(sentence, dict_type='unidic'):
  wdict_list = mecab_parse(sentence, dict_type)
  result = ''
  for wdict in wdict_list:
    word = wdict['word']
    #print(word)
    #print(type(word))
    if word==None:
      word = '<' + wdict['pos'] + '>'
    result = result + word

  return result

def mecab_parse_pos(sentence, dict_type='unidic'):
  wdict_list = mecab_parse(sentence, dict_type)
  result = ''
  for wdict in wdict_list:
    line = '<word:{},kana:{},lemma:{},pos:{},itype:{},iform:{}>'.format(
        wdict['word'],wdict['kana'],wdict['lemma'],wdict['pos'],
        wdict['itype'],wdict['iform'])
    result = result + line + '\n'

  return result
```

## 定義句型測試方法

```
import re

# 回傳句型正規式pattern套用到例句sentence的匹配結果
#   無匹配回傳None，有匹配回傳匹配物件
# def pattern_pass(pattern, sentence, dict_type='unidic'):
def pattern_pass(pattern, sentence, dict_type='ipadic'):
  pos_line = mecab_parse_pos(sentence, dict_type)
  m = re.search(pattern, pos_line)
  #print(f'pattern={pattern}\nsentence={sentence}\ndict_type={dict_type}\npos_line={pos_line}\nm=re.search(pattern,pos_line)={m}')
  return m

# 分析sentence例句，顯示套用re_pattern句型正規式的匹配結果
# def test_single_sentence(sentence, re_pattern, dict_type='unidic'):
def test_single_sentence(sentence, re_pattern, dict_type='ipadic'):
  seg_sentence = mecab_parse_pos(sentence, dict_type)

  print(f're_pattern(正規式): \n{re_pattern}\n\nseg_sentence(例句{dict_type}斷詞結果): \n{seg_sentence}')

  print('例句匹配結果: 下標數字後面，Pass表匹配，Fail表不匹配')
  mark = 'Pass' if pattern_pass(re_pattern, sentence, dict_type) else 'Fail'   # P表匹配，F表不匹配
  # marked_sentence = mark_sentence_multiple_more(sentence, re_pattern, dict_type)
  if is_multiple_pattren(re_pattern):
    marked_sentence = mark_sentence_multiple_more(sentence, re_pattern, dict_type)
  else:
    marked_sentence = mark_sentence(sentence, re_pattern, dict_type)
  print('%s %s' % (mark, marked_sentence))

# 從斷詞結果mecab_sequence，還原原始字串
def extract_word_from_mecab(mecab_sequence):
  match_words = ''
  for line in mecab_sequence.split("\n"):
    mo = re.match("<word:(.*),kana", line)
    if mo is None:
      continue
    word = mo.group(1)
    match_words += word
    # print(f'line:{line}, word:{word}')
  return match_words

# 標示sentence例句，回傳套用re_pattern句型正規式的匹配字串
def mark_sentence(sentence, re_pattern, dict_type='unidic'):
  pos_line = mecab_parse_pos(sentence, dict_type)
  sent = ''
  last = 0
  for m in re.finditer(re_pattern, pos_line):
    pre_match_pos  = pos_line[last:m.start()]
    in_match_pos   = pos_line[m.start():m.end()]
    # in_match_pos: '<word:綺麗,kana:キレイ,lemma:奇麗,pos:形状詞,一般,*,*,itype:*,iform:*>\n<word:に,kana:ニ,lemma:だ,pos:助動詞,*,*,*,itype:助動詞-ダ,iform:連用形-ニ>\n<word:畳ん,kana:タタン,lemma:畳む,pos:動詞,一>般,*,*,itype:五段-マ行,iform:連用形-撥音便>'

    last = m.end()

    pre_match_words = extract_word_from_mecab(pre_match_pos)
    pre_anchor = "{}".format(pre_match_words)

    # in_match_words: 綺麗に畳ん
    in_match_words = extract_word_from_mecab(in_match_pos)
    in_anchor = "【{}】".format(in_match_words)
    sent = sent + pre_anchor + in_anchor

  post_match_pos = pos_line[last:]
  post_match_words = extract_word_from_mecab(post_match_pos)
  post_anchor = "{}".format(post_match_words)
  sent = sent + post_anchor

  return sent

def mark_sentence_multiple(sentence, re_pattern, dict_type='unidic'):
  pos_line = mecab_parse_pos(sentence, dict_type)
  sent = ''
  last = 0
  for m in re.finditer(re_pattern, pos_line):
    pre_match_pos  = pos_line[last:m.start()]
    in_match_pos   = pos_line[m.start():m.end()]
    # in_match_pos: '<word:綺麗,kana:キレイ,lemma:奇麗,pos:形状詞,一般,*,*,itype:*,iform:*>\n<word:に,kana:ニ,lemma:だ,pos:助動詞,*,*,*,itype:助動詞-ダ,iform:連用形-ニ>\n<word:畳ん,kana:タタン,lemma:畳む,pos:動詞,一>般,*,*,itype:五段-マ行,iform:連用形-撥音便>'

    last = m.end()

    pre_match_words = extract_word_from_mecab(pre_match_pos)
    pre_anchor = "{}".format(pre_match_words)

    # in_match_words: 綺麗に畳ん
    if len(m.groups()) > 1:
      in_match_pos1 = m.group(1)
      in_match_pos3 = m.group(3)
      in_match_pos2_start = len(in_match_pos1) + 1
      in_match_pos2_end = len(in_match_pos)-len(in_match_pos3) - 1
      in_match_pos2 = in_match_pos[in_match_pos2_start:in_match_pos2_end]
      # print(f'@{in_match_pos1}@')
      # print(f'+{in_match_pos3}+')
      # print(f'#{in_match_pos2}#')
        # in_match_words: 綺麗に畳ん
      in_match_words1 = extract_word_from_mecab(in_match_pos1)
      in_match_words2 = extract_word_from_mecab(in_match_pos2)
      in_match_words3 = extract_word_from_mecab(in_match_pos3)
      in_anchor = f'【{in_match_words1}】{in_match_words2}【{in_match_words3}】'
    else:
      in_match_words = extract_word_from_mecab(in_match_pos)
      in_anchor = "【{}】".format(in_match_words)

    sent = sent + pre_anchor + in_anchor

  post_match_pos = pos_line[last:]
  post_match_words = extract_word_from_mecab(post_match_pos)
  post_anchor = "{}".format(post_match_words)
  sent = sent + post_anchor

  return sent

def mark_sentence_multiple_more(sentence, re_pattern, dict_type='unidic'):
  pos_line = mecab_parse_pos(sentence, dict_type)
  sent = ''
  last = 0
  for m in re.finditer(re_pattern, pos_line):
    pre_match_pos  = pos_line[last:m.start()]
    in_match_pos   = pos_line[m.start():m.end()]

    last = m.end()

    pre_match_words = extract_word_from_mecab(pre_match_pos)
    pre_anchor = "{}".format(pre_match_words)

    if len(m.groups()) == 4:
      in_match_pos1 = m.group(1)
      in_match_pos3 = m.group(3)
      in_match_pos2_start = len(in_match_pos1) + 1
      in_match_pos2_end = len(in_match_pos)-len(in_match_pos3) - 1
      in_match_pos2 = in_match_pos[in_match_pos2_start:in_match_pos2_end]

      in_match_words1 = extract_word_from_mecab(in_match_pos1)
      in_match_words2 = extract_word_from_mecab(in_match_pos2)
      in_match_words3 = extract_word_from_mecab(in_match_pos3)
      in_anchor = f'【{in_match_words1}】{in_match_words2}【{in_match_words3}】'
    elif len(m.groups()) == 5:
      in_match_pos1 = m.group(1)
      in_match_pos3 = m.group(3)
      in_match_pos5 = m.group(5)
      

      in_match_pos2_start = len(in_match_pos1) + 1
      in_match_pos2_end = in_match_pos.find(in_match_pos3) - 1
      in_match_pos2 = in_match_pos[in_match_pos2_start:in_match_pos2_end]

      in_match_pos4_start = in_match_pos.find(in_match_pos3) + len(in_match_pos3) + 1
      in_match_pos4_end = in_match_pos.find(in_match_pos5) - 1
      in_match_pos4 = in_match_pos[in_match_pos4_start:in_match_pos4_end]

      in_match_words1 = extract_word_from_mecab(in_match_pos1)
      in_match_words2 = extract_word_from_mecab(in_match_pos2)
      in_match_words3 = extract_word_from_mecab(in_match_pos3)
      in_match_words4 = extract_word_from_mecab(in_match_pos4)
      in_match_words5 = extract_word_from_mecab(in_match_pos5)
      in_anchor = f'【{in_match_words1}】{in_match_words2}【{in_match_words3}】{in_match_words4}【{in_match_words5}】'
    else:
      in_match_words = extract_word_from_mecab(in_match_pos)
      in_anchor = "【{}】".format(in_match_words)

    sent = sent + pre_anchor + in_anchor

  post_match_pos = pos_line[last:]
  post_match_words = extract_word_from_mecab(post_match_pos)
  post_anchor = "{}".format(post_match_words)
  sent = sent + post_anchor

  return sent

def is_multiple_pattren(re_pattern):
  if '?(<word:.*,kana:.*,lemma:.*,pos:.*,itype:.*,iform:.*>\n)*' in re_pattern:
    return True
  elif '?(<word:[^、､]*,kana:.*,lemma:.*,pos:.*,itype:.*,iform:.*>\n)*' in re_pattern:
    return True
  else:
    return False


# 針對句型代碼 source_key，分析其第sent_no個例句，顯示套用其句型正規式re_pattern的匹配結果
def test_single_source_sentence(source_key, sent_no, re_pattern, dict_type='unidic'):
  source = source_json[source_key]
  print('source_key(句型代碼): {}\nname(名稱): {}\npattern(說明): {}\n'.format(source_key, source['name'], source['pattern']))

  sents = source['sample']
  sentence = sents[sent_no]
  seg_sentence = mecab_parse_pos(sentence, dict_type)

  print(f're_pattern(正規式): \n{re_pattern}\n\nseg_sentence(例句{dict_type}斷詞結果): \n{seg_sentence}')

  print('例句匹配結果: 下標數字後面，Pass表匹配，Fail表不匹配')
  mark = 'Pass' if pattern_pass(re_pattern, sentence, dict_type) else 'Fail'   # P表匹配，F表不匹配
  if is_multiple_pattren(re_pattern):
    marked_sentence = mark_sentence_multiple_more(sentence, re_pattern, dict_type)
  else:
    marked_sentence = mark_sentence(sentence, re_pattern, dict_type)
  # marked_sentence = mark_sentence(sentence, re_pattern, dict_type)
  # marked_sentence = mark_sentence_multiple(sentence, re_pattern, dict_type)
  print('%2d %s %s' % (sent_no, mark, marked_sentence))

# 
def get_re_pattern_key(source_key, source_dict_type):
  source = source_json[source_key]

  re_pattern_key = f're_pattern_{source_dict_type}' 
  if not re_pattern_key in source: # for old corpus compatibility
    re_pattern_key = f're_pattern'

  if not re_pattern_key in source:  
    re_pattern_key = None
  
  return re_pattern_key
```
